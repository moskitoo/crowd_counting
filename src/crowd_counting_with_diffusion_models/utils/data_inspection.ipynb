```python
label_paths = [path for path in os.scandir(os.path.join(dataset_path, labels_dir))]
```

```python
labels_dir_path = os.path.join(dataset_path, labels_dir)
```

```python
numpy_label_target_directory = 'ground_truth_numpy'
```

```python
for label_path in label_paths:
    label_full_path = os.path.join(labels_dir_path, label_path)
    mat = scipy.io.loadmat(label_path)
    label = mat["image_info"][0, 0][0, 0][0]
    label_name = label_path.name
    label_name = label_name.split('.')[0] + '.npy'

    target_dir = os.path.join(dataset_path, numpy_label_target_directory)
    os.makedirs(target_dir, exist_ok=True)
    
    target_path = os.path.join(target_dir, label_name)
    np.save(target_path, label)
```

```python
images_dir_path = os.path.join(dataset_path, images_dir)
```

```python
image_paths = [path for path in os.scandir(images_dir_path)]
```

```python
shape_list = []
for image_path in image_paths:
    image_full_path = os.path.join(images_dir_path, image_path)

    image = Image.open(image_full_path)
        
    if image.mode != "RGB":
        image = image.convert("RGB")

    image = np.array(image)

    shape_list.append(image.shape)

shape_array = np.array(shape_list)
mean_shape = shape_array.mean(axis=0)

print(mean_shape)
```
